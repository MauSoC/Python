```python
#Ejercicio2
if __name__ == '__main__':
    tup = ()
    lista =[]
    aux = 0
    for i in range(10):
        num=int(input('Ingresa un numero: '))
        lista.append(num)
    tup = tuple(lista)
    mul=int(input('Indique los multiplos: '))
    if mul>0:
        for i in tup:
            if i%mul == 0:
                aux+=1
    else:
        print('EL multiplo ingresado no es correcto')

    print(f'El numero de multiplos de {mul} es: {aux}')
```

```python
#Ejercicio3
aux = 0
dic = {}
lista_h = ['Nombre', 'Telefono', 'Edad']

while True:
    opt = int(input('''¿Quieres agregar un contacto?:
                    1.-Si
                    2.-No'''))
    lis_datos = []
    if opt == 2:
        break
    for i in lista_h:
        dato = input(f'Ingrese {i}: ')
        lis_datos.append(dato)
    aux += 1
    dic[aux] = lis_datos


print(dic.items())

```

```python
#Inventar_Ejercicio 1
vocal = ['a','e','i','o','u']

letra = input('Ingrese una letra: ')
if letra.lower() in vocal:
    print(f'La letra {letra}, es una vocal')
else:
    print(f'La letra {letra}, no es una vocal')
```

```python
#Inventar_Ejercicio 2
dic = {}

for i in range(3):
    mar = input('Agregar una materia: ')
    cal = float(input('Ingrese calificacion: '))
    dic[mar]=cal

pro = sum(dic.values())/len(dic)

print(dic.items())

print(f'El promedio de las calificaciones es: {pro}')
```
